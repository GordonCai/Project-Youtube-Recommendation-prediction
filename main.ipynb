```python
video_set = data['video_id'].values
video_list = data['video_id'].values.reshape(-1,1)

for i in range(1,21,1):
    video_set = np.concatenate((video_set,data[str(i)].values),axis=0) # concatenate vertically
    video_list = np.concatenate((video_list,data[str(i)].values.reshape(-1,1)),axis=1) # concatenate horizontally
```

```python
# build hash map to speed up search
hashmap_video = {}; hashmap_cate = {}; hashmap_sim = {}; ind1=0; ind2=0
parent_video = data['video_id'].values
parent_video_set = set(parent_video)
for index, row in data.iterrows():
    hashmap_video[row['video_id']] = {'category': row['category'],'index':ind2}
    try:
        hashmap_cate[row['category']].append(row['video_id'])
    except:
        hashmap_cate[row['category']] = []
        hashmap_cate[row['category']].append(row['video_id'])
        hashmap_sim[row['category']] = ind1
        ind1 += 1
    ind2+=1
```

```python
true_edges = []

same_cate= []
cate_pair_ct = np.zeros((14,14))

# form edge pair
import math
for index0, row in enumerate(video_list):
    for i in range(20):
        try:
            math.isnan(row[i+1])
        except:
            if row[i+1] in parent_video_set:
                true_edges.append((row[0], row[i+1]))
                
                if hashmap_video[row[0]]['category'] == hashmap_video[row[i+1]]['category']:
                    same_cate.append(1)
                else: same_cate.append(0)
                    
                cate_pair_ct[hashmap_sim[hashmap_video[row[0]]['category']],hashmap_sim[hashmap_video[row[i+1]]['category']]]+=1
#                 cate_pair_ct[hashmap_sim[hashmap_video[row[i+1]]['category']],hashmap_sim[hashmap_video[row[0]]['category']]]+=1
                
# np.fill_diagonal(cate_pair_ct,0)
```

### directed graph

```python
hashmap_gtest = {}
for index, node in enumerate(g_test.nodes()):
    hashmap_gtest[node]=result[index]
```

```python
clu0_edges = []
clu1_edges = []
clu2_edges = []
clu0_nodes = []
clu1_nodes = []
clu2_nodes = []

# form edge pair
import math
for index0, row in enumerate(video_list):
    if hashmap_gtest[row[0]] == 0:
        clu0_nodes.append(row[0])
        for i in range(20):
            try:
                math.isnan(row[i+1])
            except:
                clu0_edges.append((row[0], row[i+1],2))
                clu0_edges.append((row[i+1], row[0],3))
                clu0_nodes.append(row[i+1])
                
    elif hashmap_gtest[row[0]] == 1:
        clu1_nodes.append(row[0])
        for i in range(20):
            try:
                math.isnan(row[i+1])
            except:
                clu1_edges.append((row[0], row[i+1],2))
                clu1_edges.append((row[i+1], row[0],3))
                clu1_nodes.append(row[i+1])
                
    else:
        clu2_nodes.append(row[0])
        for i in range(20):
            try:
                math.isnan(row[i+1])
            except:
                clu2_edges.append((row[0], row[i+1],2))
                clu2_edges.append((row[i+1], row[0],3))
                clu2_nodes.append(row[i+1])
```

```python
g0=nx.DiGraph()
g0.add_nodes_from(list(set(clu0_nodes))[1:])
g0.add_weighted_edges_from(clu0_edges)
len(g0.nodes())
```

```python
g1=nx.DiGraph()
g1.add_nodes_from(list(set(clu1_nodes))[1:])
g1.add_weighted_edges_from(clu1_edges)
len(g1.nodes())
```

```python
g2=nx.DiGraph()
g2.add_nodes_from(list(set(clu2_nodes))[1:])
g2.add_weighted_edges_from(clu2_edges)
len(g2.nodes())
```

```python
start=time.time()
path_length_0=np.ones(99500)*100
path_length_1=np.ones(99500)*100
path_length_2=np.ones(99500)*100

for index, pair in enumerate(test_pairs):
    if output[index] != 1:
        try:
            path_length_0[index]=nx.dijkstra_path_length(g0, source=pair[0], target=pair[1])
        except: None
            
        try:
            path_length_1[index]=nx.dijkstra_path_length(g1, source=pair[0], target=pair[1]) 
        except: None
            
        try:
            path_length_2[index]=nx.dijkstra_path_length(g2, source=pair[0], target=pair[1])
        except: None
    break
print(time.time()-start)
```
